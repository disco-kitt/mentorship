```python
# Step 1: Install necessary library
!pip install fpdf

# Step 2: Import necessary libraries
import pandas as pd
import numpy as np
import re
from fpdf import FPDF  # Ensure fpdf is installed
from google.colab import files
from IPython.display import display  # For displaying DataFrames in Colab

# Step 3: Upload CSV files
print("Please upload your CSV files for 'Students' and 'Alumni'")
uploaded = files.upload()  # Prompt file upload in Google Colab

# Automatically detect file names from uploaded files
mentee_csv = "SP2025_Students.csv"
mentor_csv = "SP2025_Alumni.csv"

for filename in uploaded.keys():
    if "Students" in filename:
        mentee_csv = filename
    elif "Alumni" in filename:
        mentor_csv = filename

# Check if both files were uploaded
if mentee_csv not in uploaded or mentor_csv not in uploaded:
    raise ValueError("Both 'Students' and 'Alumni' CSV files must be uploaded.")

# Load dataframes for mentees and mentors
mentees_df = pd.read_csv(mentee_csv, encoding="utf-8")
mentors_df = pd.read_csv(mentor_csv, encoding="utf-8")

# Display first few rows to verify data structure
print("\n🔹 Mentees Data Preview:")
display(mentees_df.head())

print("\n🔹 Mentors Data Preview:")
display(mentors_df.head())

# Function to extract company names from text
def extract_company(text):
    company_keywords = [
        "EF", "Deloitte", "Google", "Amazon", "Microsoft", "IBM",
        "Goldman Sachs", "Facebook", "Tesla", "Apple", "Accenture",
        "KPMG", "PwC", "EY"
    ]
    if pd.notna(text):
        return [company for company in company_keywords if re.search(rf"\b{company.lower()}\b", text.lower())]
    return []
```

```python
# Function to calculate match score based on updated column names
def calculate_match_score(mentee, mentor):
    score = 0

    # Ensure no NaN values in critical columns
    mentee = mentee.fillna("")
    mentor = mentor.fillna("")

    # 1. Matching based on career targets (out of 150 points)
    mentee_targets = mentee.get(
        "What specific career goals are you currently working toward? Include details such as the industry, job function, or geographic location you are targeting. How do you hope the mentorship program will help you refine or achieve these goals?",
        ""
    ).lower().strip().split()

    mentor_skills = mentor.get(
        "What specific skills, knowledge, or expertise do you feel most confident sharing with your mentee? This could include areas such as industry insights, technical skills, leadership development, or career planning.",
        ""
    ).lower().strip().split()

    if mentee_targets and mentor_skills:
        match_count = len(set(mentee_targets) & set(mentor_skills))
        score += 150 * (match_count / len(mentee_targets)) if len(mentee_targets) > 0 else 0

    # 2. Matching based on long-term goals (out of 150 points)
    mentee_goals = mentee.get(
        "What are your current career aspirations? Describe where you see yourself after graduation and how a mentor could guide you in making progress toward these goals.",
        ""
    ).lower().strip().split()

    if mentee_goals and mentor_skills:
        match_count = len(set(mentee_goals) & set(mentor_skills))
        score += 150 * (match_count / len(mentee_goals)) if len(mentee_goals) > 0 else 0

    # 3. Matching based on ideal mentor/mentee descriptions (out of 150 points)
    mentee_description = mentee.get(
        "What qualities or expertise would your ideal mentor have? Describe the type of mentor who could best support your growth, including their background, skills, or experiences that would align with your goals.",
        ""
    ).lower().strip().split()

    mentor_description = mentor.get(
        "What qualities or characteristics do you think make for an ideal mentee? Describe the type of student you believe you can support most effectively, including their goals, personality, or areas of interest.",
        ""
    ).lower().strip().split()

    if mentee_description and mentor_description:
        match_count = len(set(mentee_description) & set(mentor_description))
        score += 150 * (match_count / len(mentee_description)) if len(mentee_description) > 0 else 0

    # 4. Company match criterion (50 points max)
    mentee_companies = extract_company(mentee.get(
        "What specific career goals are you currently working toward? Include details such as the industry, job function, or geographic location you are targeting. How do you hope the mentorship program will help you refine or achieve these goals?",
        ""
    ))

    mentor_companies = extract_company(mentor.get(
        "Current Company (If you're a student, please list your current university.)",
        ""
    ))

    if len(set(mentee_companies) & set(mentor_companies)) > 0:
        score += 50  # Company match bonus

    # 5. Matching based on major (out of 100 points)
    if mentee.get("Major", "").lower().strip() == mentor.get("What was your major at Hult?", "").lower().strip():
        score += 100  # Full score for matching majors

    # 6. Matching based on preferred meeting time (out of 25 points)
    mentee_times = set(mentee.get(
        "What are your preferred times for scheduling mentorship meetings? (Select all that apply.)",
        ""
    ).lower().strip().split(', '))

    mentor_times = set(mentor.get(
        "What times of day are most convenient for you to schedule mentorship meetings? Select all options that apply.",
        ""
    ).lower().strip().split(', '))

    if mentee_times and mentor_times:
        match_count = len(mentee_times & mentor_times)
        score += 25 * (match_count / len(mentee_times)) if len(mentee_times) > 0 else 0

    # 7. Matching based on preferred communication method (out of 25 points)
    mentee_communication = set(mentee.get(
        "What are your preferred methods for communicating with your mentor? (Select all that apply.)",
        ""
    ).lower().strip().split(', '))

    mentor_communication = set(mentor.get(
        "How do you prefer to communicate with your mentee? Please select all methods that work best for you.",
        ""
    ).lower().strip().split(', '))

    if mentee_communication and mentor_communication:
        match_count = len(mentee_communication & mentor_communication)
        score += 25 * (match_count / len(mentee_communication)) if len(mentee_communication) > 0 else 0

    return score

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Define the mentees and mentors based on the data in their respective DataFrames
mentees = mentees_df['Name (First)'] + ' ' + mentees_df['Name (Last)']
mentors = mentors_df['Name (First)'] + ' ' + mentors_df['Name (Last)']

# Step 4: Initialize the compatibility matrix (mentees as rows and mentors as columns)
compatibility_matrix = pd.DataFrame(0, index=mentees, columns=mentors)

# Step 5: Remove duplicate columns from the mentor columns if present
compatibility_matrix = compatibility_matrix.loc[:, ~compatibility_matrix.columns.duplicated()]

# Step 6: Calculate and populate the compatibility matrix using the match score function
for mentee_idx, mentee_row in mentees_df.iterrows():
    for mentor_idx, mentor_row in mentors_df.iterrows():
        try:
            # Calculate match score for each mentee-mentor pair
            score = calculate_match_score(mentee_row, mentor_row)
        except KeyError as e:
            # Handle missing columns gracefully
            print(f"Error: Missing column {e} in the mentor or mentee data. Please check the CSV files.")
            continue  # Skip this pair if there is a missing key

        compatibility_matrix.at[
            f"{mentee_row['Name (First)']} {mentee_row['Name (Last)']}",
            f"{mentor_row['Name (First)']} {mentor_row['Name (Last)']}"
        ] = score

# Step 7: Normalize the compatibility matrix to scale the scores
normalized_matrix = compatibility_matrix.div(compatibility_matrix.max(axis=1), axis=0).fillna(0)

# Step 8: Visualize the compatibility matrix using a heatmap (non-normalized)
plt.figure(figsize=(24, 18))
sns.heatmap(compatibility_matrix, cmap="coolwarm", annot=True, fmt=".1f", linewidths=.5, annot_kws={"size": 10})
plt.title('Mentee-Mentor Compatibility Heatmap (Non-Normalized)')
plt.xticks(rotation=65, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Save the heatmap (non-normalized)
heatmap_filename_raw = "compatibility_heatmap_raw.png"
plt.savefig(heatmap_filename_raw)
plt.show()

# Step 9: Visualize the normalized compatibility matrix using a heatmap (normalized)
plt.figure(figsize=(24, 18))
sns.heatmap(normalized_matrix, cmap="coolwarm", annot=True, fmt=".2f", linewidths=.5, annot_kws={"size": 10})
plt.title('Mentee-Mentor Compatibility Heatmap (Normalized)')
plt.xticks(rotation=65, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Save the heatmap (normalized)
heatmap_filename_normalized = "compatibility_heatmap_normalized.png"
plt.savefig(heatmap_filename_normalized)
plt.show()
```
